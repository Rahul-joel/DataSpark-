```python
#Geographical Analysis: Analyze sales by store location to identify high-performing regions.
query10=''' select StoreKey,store_country,store_state,sum(`Total Sales USD`) as total_sales 
from dataspark.spark_dataset 
group by StoreKey,store_country,store_state 
order by total_sales desc'''

Geographical_Analysis=pd.read_sql_query(query10,connect)
Geographical_Analysis
```

```python
# Category Analysis: Analyze sales performance across different product categories and subcategories.
query9=''' SELECT 
    Category, 
    Subcategory, 
    `Product Name`, 
    SUM(Quantity) AS total_quantity_sold, 
    SUM(`Total Sales USD`) AS total_sales 
FROM 
    dataspark.spark_dataset 
GROUP BY 
    Category, 
    Subcategory, 
    `Product Name`
'''

Category_Analysis=pd.read_sql_query(query9,connect)
Category_Analysis
```

```python
#Profitability Analysis: Calculate profit margins for products by comparing unit cost and unit price.
query8 = ''' 
SELECT 
    ProductKey, 
    `Product Name`,
    `Unit Cost USD`, 
    `Unit Price USD`, 
    SUM(Quantity * (`Unit Price USD` - `Unit Cost USD`)) AS total_profit,
    (SUM(Quantity * (`Unit Price USD` - `Unit Cost USD`)) / SUM(Quantity * `Unit Price USD`) * 100) AS profit_margin
FROM 
    dataspark.spark_dataset 
GROUP BY 
    ProductKey, `Product Name`, `Unit Cost USD`, `Unit Price USD` 
'''

Profitability_Analysis = pd.read_sql_query(query8, connect)
Profitability_Analysis
```

```python
#Product Popularity: Identify the most and least popular products based on sales data.
query7 = ''' 
SELECT 
    ProductKey, 
    `Product Name`, 
    SUM(Quantity) AS total_quantity, 
    `Unit Price USD`,
    SUM(`Total Sales USD`) AS total_sales 
FROM 
    dataspark.spark_dataset 
GROUP BY 
    ProductKey, `Product Name`, `Unit Price USD` 
ORDER BY 
    total_quantity DESC, total_sales DESC
'''

product_popularity = pd.read_sql_query(query7, connect)
product_popularity
```

```python
#Sales by Currency: Examine how different currencies impact sales figures, considering exchange rates.

query6 = ''' 
SELECT 
    DATE_FORMAT(`Order Date`, '%%Y-%%m-%%d') AS order_month, 
    `Currency Code`,
    `Exchange`, 
    SUM(`Total Sales USD`) AS total_sales_usd,
    SUM(`Total Sales USD` * `Exchange`) AS total_sales_currency
FROM 
    dataspark.spark_dataset 
GROUP BY 
    order_month, 
    `Currency Code`, 
    `Exchange`
ORDER BY 
    order_month
    
'''

sales_currency = pd.read_sql_query(query6, connect)
sales_currency
```

```python
#Sales by Store: Assess the performance of different stores based on sales data.
query5 = ''' 
SELECT 
    StoreKey,`Square Meters`,
    SUM(`Total Sales USD`) AS total_revenue 
FROM 
    dataspark.spark_dataset 
GROUP BY 
    StoreKey,`Square Meters`
ORDER BY 
    total_revenue DESC;
'''

Sales_by_Store = pd.read_sql_query(query5, connect)
Sales_by_Store
```

```python
#Sales by Product: Evaluate which products are the top performers in terms of quantity sold and revenue generated.
query4 = ''' 
SELECT 
    DATE_FORMAT(`Order Date`, '%%Y-%%m') AS order_month, 
    `Product Name`, 
    SUM(Quantity) AS total_quantity,      -- Summing the quantities sold
    SUM(`Total Sales USD`) AS total_revenue  -- Summing the revenue generated
FROM 
    dataspark.spark_dataset 
GROUP BY 
    order_month, 
    `Product Name`  -- Also group by product name to evaluate performance by product
ORDER BY 
    order_month asc,total_quantity DESC;  -- Order by total quantity sold to get top performers
'''

Sales_by_Product = pd.read_sql_query(query4, connect)
Sales_by_Product
```

```python
#Overall Sales Performance: Analyze total sales over time, identifying trends and seasonality
query3 = ''' 
SELECT 
    `customer_country`, 
    `customer_state`, 
    DATE_FORMAT(`Order Date`, '%%Y-%%m') AS order_month, 
    SUM(`Total Sales USD`) AS total_sales 
FROM 
    dataspark.spark_dataset 
GROUP BY 
    `customer_country`, `customer_state`, order_month 
ORDER BY 
    order_month;
'''

sales_performance = pd.read_sql_query(query3, connect)
sales_performance
```

```python
#Purchase Patterns: Identify purchasing patterns such as average order value, frequency of purchases, and preferred products.
query2 = ''' 
SELECT 
    Name AS customer_name, 
    `Product Name`, 
    COUNT(`Order Number`) AS purchase_frequency, 
    AVG(`Total Sales USD`) AS avg_order_value 
FROM 
    dataspark.spark_dataset 
GROUP BY 
    Name, `Product Name` 
ORDER BY 
    purchase_frequency DESC;
'''

customer_purchase_frequency = pd.read_sql_query(query2, connect)
customer_purchase_frequency
```

```python
# 1.Demographic Distribution: Analyze the distribution of customers based on gender, age,city, state, country, continent.
query1=''' select  Continent,customer_country,customer_state,CASE 
        WHEN Age BETWEEN 18 AND 25 THEN '18-25'
        WHEN Age BETWEEN 26 AND 35 THEN '26-35'
        WHEN Age BETWEEN 36 AND 45 THEN '36-45'
        WHEN Age BETWEEN 46 AND 55 THEN '46-55'
        ELSE '55+' 
      END AS age_group,Gender,count(*) as customer_count from dataspark.spark_dataset group by Continent,customer_country,customer_state,age_group,Gender order by customer_count desc'''
Demographic_Analysis=pd.read_sql_query(query1,connect)
Demographic_Analysis

```
